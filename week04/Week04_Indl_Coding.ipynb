```python
df = pd.DataFrame(grocery, columns = ['Location'])
df
```

```python
df.sample(15)
```

```python
lat = []
lon = []
```

```python
for row in df['Location']: 
    try: 
        lat.append(row.split(',')[0]) 
        lon.append(row.split(',')[1])
    except:
        lat.append(np.NaN)
        lon.append(np.NaN)
```

```python
df['latitude'] = lat
df['longitude'] = lon
```

```python
df
```

```python
df['latitude']= df['latitude'].str.replace("(","")
```

```python
df['latitude'].head()
```

```python
df['longitude']= df['longitude'].str.replace(")","")
```

```python
df['longitude'].head()
```

```python
latitude = (df.latitude).apply(lambda x: float(x))
latitude
```

```python
longitude = (df.longitude).apply(lambda x: float(x))
longitude
```

```python
latitude.dtypes
```

```python
longitude.dtypes
```

```python
gdf = gpd.GeoDataFrame(
    df, geometry=gpd.points_from_xy(longitude, latitude))
```

```python
gdf = gpd.GeoDataFrame(df, 
                       crs='epsg:4326',
                       geometry=gpd.points_from_xy(gdf.longitude, gdf.latitude))
```

```python
gdf_web_mercator = gdf.to_crs(epsg=3857)
```

```python
fig, ax = plt.subplots(figsize=(30,30))

gdf_web_mercator.plot(ax=ax, marker='*', color='green')

ax.axis('off')

ax.set_title('Grocery Stores in West LA & South Central LA')

ctx.add_basemap(ax)
```
