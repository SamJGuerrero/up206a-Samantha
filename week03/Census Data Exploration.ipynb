## Sorting
I can sort the data by their respective columns. Since I am trying to pull the **pe_description** of foodmkt retail only, I am going to focus on this column.

```python
gdf_sorted = gdf.sort_values(by='pe_description')
```

```python
gdf_sorted[['facility_name','pe_description','facility_address','facility_city','facility_state',
 'facility_zip','facility_latitude','facility_longitude','geometry']].head(299)
```

Using *head.()*, I pulled only the food market retail category. Now I can plot these!

```python
gdf_sorted.head(299).plot()
```

This plot is cute and will do, but I can add a lot more characteristics. 

```python
gdf_sorted.head(299).plot(figsize=(10,10),column='pe_description',legend=True)
```

Now, we have all food market retails that are labelled by high, low, and moderate risk.

## More detailed maps!

Thanks to geopandas, I can create a choropleth map.

```python
gdf_sorted.head(299).plot(figsize=(12,10),
                 column='pe_description',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
gdf_sorted.head(299).plot(figsize=(12,10),
                 column='pe_description',
                 legend=True, 
                 scheme='equal_interval')
```

```python
gdf_sorted.head(299).plot(figsize=(12,10),
                 column='pe_description',
                 legend=True, 
                 scheme='quantiles')
```

Since the range in our legend is not a high number, there is not enough data to create many quartiles, which gives us the error message above.

## Add a basemap

I am going to attempt to add a basemap by reprojecting my geodataframe to Web Mercator (epsg: 3857).

```python
gdf_web_mercator = gdf_sorted.head(299).to_crs(epsg=3857)
```

```python
fig, ax = plt.subplots(figsize=(15, 15))

gdf_web_mercator[gdf_web_mercator['pe_description']].plot(ax=ax, alpha=0.8)

ax.axis('off')

ax.set_title('Food Description',fontsize=16)

ctx.add_basemap(ax)
```
